```python
query = 'Ask HN: Who is hiring?'

import datetime


created_at_i = datetime.datetime(2025, 5, 1, 0, 0, 0, tzinfo=datetime.timezone.utc).timestamp()

encoded_query = {
    'query': query,
    'tags': 'ask_hn,author_whoishiring',
    'numericFilters': f'created_at_i>{created_at_i}',
}

api_url = f'https://hn.algolia.com/api/v1/search_by_date'

response = httpx.get(api_url, params=encoded_query)
data = response.json()
```

```python
story_id = data['hits'][0]['story_id']
httpx.get('https://hn.algolia.com/api/v1/items/43858554').json()['children'][::-1]
```
